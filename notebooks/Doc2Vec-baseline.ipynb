```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
```

```python
# Download required resources for NLTK
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Define stopwords and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Preprocess function
def preprocess_text(text):
    # Remove non-alphanumeric characters and convert to lowercase
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text.lower())
    # Tokenize the text into words
    tokens = word_tokenize(text)
    # Remove stopwords and lemmatize the words
    preprocessed_tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
    # Join the preprocessed tokens back into a single string
    preprocessed_text = ' '.join(preprocessed_tokens)
    return preprocessed_text
```

```python
df['Combined_Review'] = df['Combined_Review'].apply(preprocess_text)
```

```python
df = pd.read_csv('/content/drive/MyDrive/clean_df.csv')
```

```python

import re

# Convert the column to string type
df['combined_review'] = df['combined_review'].astype(str)

# Remove numbers from the text in the specific column
df['combined_review'] = df['combined_review'].apply(lambda x: re.sub(r'\d+', '', x))

# Display the updated DataFrame
print(df)
```

```python
# Create TaggedDocuments from the text in the specific column

documents = [TaggedDocument(words=text.split(), tags=[i]) for i, text in enumerate(df['combined_review'])]
```

```python
documents
```

```python
# Train the Doc2Vec model
model = Doc2Vec(documents, vector_size=100, window=5, min_count=1, workers=4)
```

```python
#Infer the document vectors for the specific column
vectors = [model.infer_vector(text.split()) for text in df['combined_review']]
```

```python
for i in range(len(vectors[0])):
    df[f'vector_{i+1}'] = [vector[i] for vector in vectors]
```
